```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
```

```python
#Plots
#Line - X-axis in order
#Bar - categorical/ordinal data
#Scatter - Relationship b/w two numerical variables / correlation
#Box(Whisker) - statistical graph of numerical variables/ compare data of different categories
#Histograms - compare frequency distributions
```

```python
#Generating range from -2 to 2
a= np.linspace(-2*np.pi,2*np.pi)
sine=np.sin(a)
cose=np.cos(a)
```

```python
plt.plot(a,sine)
plt.plot(a,cose)
plt.show()
```

```python
#use plt.figure to give size-attributes to plot
plt.figure(figsize=(8,8),dpi=120)
plt.subplot(111)
plt.plot(a,sine,color="red",linewidth=1,linestyle="--")
plt.plot(a,cose,color="blue",linewidth=2.5,marker='o',linestyle="-")
plt.xlim(a.min()*1.1,a.max()*1.1)
plt.ylim(cose.min()*1.1,cose.max()*1.1)
plt.title("Sine and Cosine")
plt.show()
```

```python
b=np.random.randn(1000)
plt.hist(b)
plt.show()
```

```python
c=np.random.rand(1000)
d=5.25*c**5.4+2.3*c+3.2*np.random.rand(1000)
plt.scatter(c,d)
plt.show()
```

```python
plt.bar(a,sine, label="Sin(a)")
plt.bar(a,cose,label="Cos(a)")
plt.legend(loc="lower left")
plt.show()
```

```python
plt.boxplot([a,b,c,d,sine,cose],labels=["a","b","c","d","sine","cose"])
plt.show()
```

```python
#For stacking bar chart define bottom=
x=["d","sine","c"]
means=[d.mean(),sine.mean(),c.mean()]
maxs=[d.max(),sine.max(),c.max()]
plt.bar(x,means,width=0.5,label="Means")
plt.bar(x,maxs,width=0.5,label="Max",bottom=means)
plt.show()
```

```python
sns.boxplot(data=[a,b,c,d,sine,cose])
plt.show()
sns.boxplot(data=[a,b,c,d])
```

```python
sns.barplot(data=[a,b,c,d,sine,cose])
plt.show()
```

```python
sns.histplot(sine)
plt.show()
```
